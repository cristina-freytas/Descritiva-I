#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
# importando 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline      


# carregando arquivo e construindo DataFrame
arquivo = pd.read_csv('sinasci_RO_2019.csv')
df_sinasci = pd.DataFrame(arquivo)

# informações sobre os dados
df_sinasci.info()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Gráfico em barras 'APGAR1'
plt.close('all')
plt.rc('figure', figsize=(8, 6))
fig, axes = plt.subplots(1, 1)

df_sinasci['APGAR1'].value_counts().sort_index().plot.bar()
axes.set_xlabel('APGAR1', color='red')
axes.set_ylabel('Frequência', color='red')
axes.set_title('Frequência de APGAR1', color='red', fontsize=16)
axes.grid(True, linestyle=':', alpha=0.3)
plt.show()
```

```python
# Gráfico em barras 'APGAR5'
plt.close('all')
plt.rc('figure', figsize=(8, 6))
fig, axes = plt.subplots(1, 1)

df_sinasci['APGAR5'].value_counts().sort_index().plot.bar()
axes.set_xlabel('APGAR5', color='red')
axes.set_ylabel('Frequência', color='red')
axes.set_title('Frequência de APGAR5', color='red', fontsize=16)
axes.grid(True, linestyle=':', alpha=0.3)
plt.show()
```

```python
# Gráfico em barras 'SEXO'
plt.close('all')
plt.rc('figure', figsize=(8, 6))
fig, axes = plt.subplots(1, 1)

df_sinasci['SEXO'].value_counts().sort_index().plot.bar()
axes.set_xlabel('Sexo', color='red')
axes.set_ylabel('Frequência', color='red')
axes.set_title('Frequência por Sexo', color='red', fontsize=16)
axes.grid(True, linestyle=':', alpha=0.3)
plt.show()

```

```python
# Gráfico em barras 'GESTACAO'

plt.close('all')
plt.rc('figure', figsize=(8, 6))
fig, axes = plt.subplots(1, 1)

df_sinasci['GESTACAO'].value_counts().sort_index().plot.bar()
axes.set_xlabel('Tempo de Gestação', color='red')
axes.set_ylabel('Frequência', color='red')
axes.set_title('Frequência por tempo de Gestação', color='red', fontsize=16)
axes.grid(True, linestyle=':', alpha=0.3)
plt.show()

```

```python
# Gráfico em barras 'ESTCIVMAE'

plt.close('all')
plt.rc('figure', figsize=(8, 6))
fig, axes = plt.subplots(1, 1)


estado_civil_mapping = {1: 'Solteira', 2: 'Casada', 3:'Viúva', 4:'Separada', 5: 'União Concensual', 9:'Ignorado'}
df_sinasci['ESTCIVMAE'].value_counts().sort_index().plot.bar()

axes.set_xlabel('Estado Civil da Mãe', color='red')
axes.set_ylabel('Frequência', color='red')
axes.set_title('Frequência por Estado Civil da Mãe', color='red', fontsize=16)
axes.grid(True, linestyle=':', alpha=0.3)

valores_unicos = df_sinasci['ESTCIVMAE'].unique()
rotulos = [estado_civil_mapping.get(valor, str(valor)) for valor in valores_unicos]

plt.xticks(range(len(valores_unicos)), rotulos)

plt.show()

```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(1, 1)

sns.barplot(x='APGAR5', y='CONSULTAS', data = df_sinasci)
axes.set_xlabel('APGAR5', color='red')
axes.set_ylabel('Número de Consultas', color='red')
axes.set_title('APGAR5 Médio por Consultas', color='red', fontsize=16)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Os gráficos funcionam dentro da lógica do matplotlib
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=df_sinasci)
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('Frequência', color='red')
axes[0, 0].set_title('Frequência por Sexo', color='red')

sns.countplot(ax = axes[0, 1], x='GESTACAO', data=df_sinasci)
axes[0, 1].set_xlabel('')
axes[0, 1].set_ylabel('Frequência', color='red')
axes[0, 1].set_title('Frequência por Gestação', color='red')
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=90, fontsize=8)


df_sinasci_mapped = df_sinasci.copy()
df_sinasci_mapped['ESTCIVMAE'] = df_sinasci_mapped['ESTCIVMAE'].map(estado_civil_mapping)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=df_sinasci_mapped, order=estado_civil_mapping.values())
axes[1, 0].set_xlabel('')
axes[1, 0].set_ylabel('Frequência', color='red')
axes[1, 0].set_title('Frequência por Estado Civil da Mãe', color='red')
axes[1, 0].set_xticklabels(estado_civil_mapping.values(), rotation=90, fontsize=8)
estado_civil_mapping = {1: 'Solteira', 2: 'Casada', 3:'Viúva', 4:'Separada', 5: 'União Concensual', 9:'Ignorado'}
valores_unicos = df_sinasci['ESTCIVMAE'].unique()
rotulos = [estado_civil_mapping.get(valor, str(valor)) for valor in valores_unicos]
plt.xticks(range(len(valores_unicos)), rotulos)



sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df_sinasci)
axes[1, 1].set_xlabel('')
axes[1, 1].set_ylabel('Frequência', color='red')
axes[1, 1].set_title('Frequência por Consultas Pré-Natal', color='red')

#ajustar o espaçamento
plt.subplots_adjust(wspace=0.5, hspace=1)
```
